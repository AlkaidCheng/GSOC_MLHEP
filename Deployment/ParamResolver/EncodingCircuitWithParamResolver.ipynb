```
from quple.data_encoding.second_order_expansion import SecondOrderExpansion
import cirq
import numpy as np
```

```
n_qubit = 5
# define encoding circuit
cq = SecondOrderExpansion(feature_dimension=n_qubit)
# print the circuit
cq
```

```
# print out the parameters in the circuit
cq.parameter_table.params
```

```
# generate 10 sets of random data
data = np.random.rand(10, n_qubit)
print(data)
```

```
# get the circuit with the parameter values resolved
resolved_circuits = cq.resolve_parameters(data)
```

```
# print out the first resolved circuit
resolved_circuits[0]
```

```
# Optional: get parameter resolvers
# for simulator.run(cq, resolver)
param_resolvers = cq.get_parameter_resolver(data)
print(param_resolvers)
```

```
# create measurements on first qubit
cq.measure(0)
```

```
#create simulator
simulator = cirq.Simulator()
```

```
# output result of measurement
res = simulator.run(cq, param_resolvers[0], repetitions=500)
print(res)
```
