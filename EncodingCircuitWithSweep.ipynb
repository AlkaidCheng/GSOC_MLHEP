```
from quple.data_encoding.second_order_expansion import SecondOrderExpansion
import numpy as np
import cirq
```

```
n_qubit = 5
# define encoding circuit
cq = SecondOrderExpansion(feature_dimension=n_qubit)
# construct circuit
cq.assemble()
# measure the last qubit
cq.append(cirq.measure(cq.qubits[n_qubit-1], key="z"))
# print the circuit
print(cq)
```

```
# print out the parameters in the circuit
cq.parameter_table.params
```

```
# generate 10 sets of random data
data = np.random.rand(10, n_qubit)
print(data)
```

```
#flatten parameters and feed data
new_cq, param_sweep = cq.flatten_parameters(data)
```

```
print(new_cq)
```

```
print(param_sweep)
```

```
simulator = cirq.Simulator()
```

```
res = simulator.run(new_cq, param_sweep[0], repetitions=500)
```

```
print(res)
```
